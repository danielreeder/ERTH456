```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
rng = np.random.default_rng(123)

from numpy import pi
from matplotlib.ticker import FuncFormatter, MultipleLocator
```

```python
####################################
#              1 ii.               #
####################################

# returns the value of the square wave for the corresponding values of t and T
def square_wave(t, T):
    t = t % T # get the current time in relation to the period
    
    if 0 < t and t < T/2:
        return 1
        
    if T/2 < t and t < T:
        return -1
        
    if t == 0 or t == T:
        return 0
        
    # return -1

# returns the Fourier approximation of the square wave for given values of t, T, and nterms
def square_wave_fourier(t, T, nterm):
    omega = (2 * pi) / T
    factor = 4 / pi
    
    agg = sum([np.sin(i * omega * t) / i for i in np.arange(1, nterm, 2)])

    return factor * agg  
  
T = 2 * pi
nsamp = 1_000

# creates an array of all time values to be used and an empty array to store wave values in
interval = np.linspace(0, 10*pi, 10*nsamp)
data_wave = np.zeros(len(interval))

# calculates wave value at each value of t between 0 and 10pi
i = 0
for t in interval:
    data_wave[i] = square_wave(t, T)
    i += 1
```

```python
####################################
#              1 iii.              #
####################################

# an array of the different values of nterm to be used
nterm_arr = np.array([5, 10, 20, 100])

# creates an array containing an empty array for each value of nterm
data_fourier = np.zeros(len(nterm_arr), dtype=np.ndarray)
for i in np.arange(len(nterm_arr)):
    data_fourier[i] = np.zeros(len(interval))

# calculates fourier approximations for each number of terms
i = 0
for nterm in nterm_arr:
    j = 0
    for t in interval:
        data_fourier[i][j] = square_wave_fourier(t, T, nterm)
        j += 1
    i += 1
        
fig, axes = plt.subplots(2, 2, figsize=(15, 15))

# plots the wave versus the fourier approximation with each number of terms
i = 0
for _ in axes:
    for ax in _:
        ax.set_title(f"Wave vs. Approximation for nterms={nterm_arr[i]}")
        ax.set_xlabel("Time")
        ax.set_ylabel("Value")
        ax.plot(interval, data_wave)
        ax.plot(interval, data_fourier[i])
        ax.xaxis.set_major_formatter(FuncFormatter(
            lambda val,pos: '{:.0f}$\pi$'.format(val/np.pi) if val !=0 else '0'
        ))
        ax.xaxis.set_major_locator(MultipleLocator(base=np.pi))
        ax.axhline(0, c='black')
        ax.set_yticks([-1, 0, 1])
        i += 1
plt.subplots_adjust(wspace=0.3, hspace=0.3)

# calculates the misfit between the wave and approximation using root mean square error
def misfit(wave, fourier):
    return np.sqrt(np.mean((wave - fourier) ** 2))

misfits = np.zeros(4)

for i in np.arange(4):
    misfits[i] = misfit(data_wave, data_fourier[i])

print(f"The Fourier series is very effective at representing the square wave in general.\nThe fit is worst at the points where discontinuities occur, which can be seen at every multiple of pi.\nThe largest misfit is {misfits[0]:.3f} and occurs when nterms is smallest. The misfit decreases substantially as nterms is increased, with the smallest misfit occuring at nterms=100: {misfits[3]:.3f}.")
```

```python
####################################
#             1 iiii.              #
####################################

# sets sampling rate and number of terms to 100
nsamp = 100; nterm = 100

interval = np.linspace(0, 10*pi, 10*nsamp)
wave_100 = np.zeros(len(interval))
fourier_100 = np.zeros(len(interval))

fig, axes = plt.subplots(2, 1, figsize=(5, 10))
                         
i = 0
for t in interval:
    wave_100[i] = square_wave(t, T)
    fourier_100[i] = square_wave_fourier(t, T, nterm)
    i += 1

axes[0].plot(interval, wave_100)
axes[0].plot(interval, fourier_100)
axes[0].set_title("Wave vs. Approximation with nsamp=100; nterm=100")
axes[0].xaxis.set_major_formatter(FuncFormatter(
            lambda val,pos: '{:.0f}$\pi$'.format(val/np.pi) if val !=0 else '0'
        ))
axes[0].xaxis.set_major_locator(MultipleLocator(base=np.pi))
axes[0].axhline(0, c='black')
axes[0].set_yticks([-1, 0, 1])

# sets sampling rate to 10,000, number of terms stays at 100
nsamp = 10_000
interval = np.linspace(0, 10*pi, 10*nsamp)

wave_10000 = np.zeros(len(interval))
fourier_10000 = np.zeros(len(interval))

i = 0
for t in interval:
    wave_10000[i] = square_wave(t, T)
    fourier_10000[i] = square_wave_fourier(t, T, nterm)
    i += 1

axes[1].plot(interval, wave_10000)
axes[1].plot(interval, fourier_10000)
axes[1].set_title("Wave vs. Approximation with nsamp=10,000; nterm=100")
axes[1].xaxis.set_major_formatter(FuncFormatter(
            lambda val,pos: '{:.0f}$\pi$'.format(val/np.pi) if val !=0 else '0'
        ))
axes[1].xaxis.set_major_locator(MultipleLocator(base=np.pi))
axes[1].axhline(0, c='black')
axes[1].set_yticks([-1, 0, 1]);
```

```python
print(f"Misfit with nsamp=100: {misfit(wave_100, fourier_100):.3f}")
print(f"Misfit with nsamp=10,000: {misfit(wave_10000, fourier_10000):.3f}")
print("The misfit with nsamp=100 is slightly smaller (numerically) than with nsamp=10,000. Visually, it seems that this is because with a higher sampling rate the errors in the approximation that occur at discontinuities are exaggerated.")
```
