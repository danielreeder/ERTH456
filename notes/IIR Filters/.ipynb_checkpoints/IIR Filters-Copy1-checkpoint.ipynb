```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib notebook
```

```python
# define filter
fc = 5
wc = 2 * np.pi * fc
b, a = signal.butter(4, wc, 'low', analog=True)

# get frequency response
w, h = signal.freqs(b, a)

# convert from angular to linear freq
f = w / (2 * np.pi)

b2, a2 = signal.butter(50, wc, 'low', analog=True)
w2, h2 = signal.freqs(b2, a2)
f2 = w2 / (2 * np.pi)
```

```python
plt.semilogx(f, np.abs(h), label="4 pole filter")
plt.semilogx(f2, np.abs(h2), label="2 pole filter")
plt.title("Butterworth filter frequency response")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(which="both")
plt.legend();
```
