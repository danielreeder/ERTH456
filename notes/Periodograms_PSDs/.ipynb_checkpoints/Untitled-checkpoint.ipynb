```python
from scipy import signal as s
from scipy.fftpack import fft, fftfreq
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from numpy import sin, pi
```

```python
f1 = 2
f2 = 5
dt = 0.005
t = np.arange(0, 10, dt)

y1 = sin(2 * pi * f1 * t)
y2 = sin(2 * pi * f2 * t)
y3 = y1 + y2
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

ax1.plot(t, y1, c='b')
ax1.plot(t, y2, c='r');
ax2.plot(t, y3, c='g')
```

```python
# take autocorrelations
y1_auto_corr = s.correlate(y1, y1, mode='full')
t_corr = np.arange(-max(t), max(t), dt)

plt.plot(t_corr, y1_auto_corr)
plt.xlabel("Lag (s)")
plt.grid(True)
plt.xticks(np.arange(-10, 11, 0.5))
plt.xlim(-3, 3);
```

```python
# take autocorrelations
y2_auto_corr = s.correlate(y2, y2, mode='full')
t_corr = np.arange(-max(t), max(t), dt)

plt.subplots(figsize=(15, 5))
plt.plot(t_corr, y2_auto_corr)
plt.xlabel("Lag (s)")
plt.grid(True)
plt.xticks(np.arange(-10, 11, 0.5))
plt.xlim(-3, 3);
```

```python
# take autocorrelations
y3_auto_corr = s.correlate(y3, y3, mode='full')
t_corr = np.arange(-max(t), max(t), dt)

plt.subplots(figsize=(15, 10))
plt.plot(t_corr, y3_auto_corr, c='b')
plt.xlabel("Lag (s)")
plt.grid(True)
plt.xticks(np.arange(-10, 11, .5));
plt.xlim(-3, 3);
```

```python
fft_corr = fft(y3_auto_corr)
freq = fftfreq(len(y3_auto_corr), dt)

plt.semilogx(freq[np.where(freq>0)], 2 * abs(fft_corr)[np.where(freq>0)])
plt.xlim(1, 6)
```
